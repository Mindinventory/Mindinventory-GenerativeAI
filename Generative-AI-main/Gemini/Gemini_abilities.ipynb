## Basic Chain

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain.prompts import ChatPromptTemplate
from langchain.schema.output_parser import StrOutputParser

prompt= ChatPromptTemplate.from_template(
  '''
    Describe the stats of {topic}
  '''
)
output_parser= StrOutputParser()
chain= prompt | llm | output_parser
Markdown(chain.invoke({"topic": "superman"}))
```

## More Complicated Chain

```python
from langchain.vectorstores import DocArrayInMemorySearch

embeddings= GoogleGenerativeAIEmbeddings(model= "models/embedding-001",
                                         google_api_key= "API")
vectorstore= DocArrayInMemorySearch.from_texts(
    [
        "Original poster: The person who started a discussion thread or made the first comment on a social media platform",
        "Overpowered: A term used in gaming to describe something that is too strong or unfair.",
        "Opinion: A term used in legal documents to refer to a statement or judgment by a court or an authority."

    ], embedding= embeddings
)
retriever= vectorstore.as_retriever()
retriever.get_relevant_documents("What is OP?")
```

```python
retriever.get_relevant_documents("What is an opinion?")
```

```python
template= '''
Answer the following question only based on the following context: {context}
Question: {question}
Return your answer as JSON
'''
prompt= ChatPromptTemplate.from_template(template)
```

```python
from langchain.schema.runnable import RunnableMap
chain= RunnableMap({
    "context": lambda x: retriever.get_relevant_documents(x["question"]),
    "question": lambda x: x["question"]
}) | prompt | llm | output_parser
```

```python
chain.invoke({"question": "What is the synonym for the word 'too strong'?"})
```

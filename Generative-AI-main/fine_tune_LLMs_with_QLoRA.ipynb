```python
import torch
import transformers
import peft
from datasets import load_dataset
```

```python
bnb_config= transformers.BitsAndBytesConfig(
    load_in_4bit= True,
    bnb_4bit_use_double_quant= True,
    bnb_4bit_quant_type= "nf4",
    bnb_4bit_compute_dtype= torch.bfloat16
)
```

```python
tokenizer= transformers.AutoTokenizer.from_pretrained("EleutherAI/gpt-neox-20b")
model= transformers.AutoModelForCausalLM.from_pretrained(
    "EleutherAI/gpt-neox-20b",
    quantization_config= bnb_config,
    device_map= {"": 0}
)
```

```python
model.gradient_checkpointing_enable()
model= peft.prepare_model_for_kbit_training(model)
```

```python
def print_trainable_parameters(model):
  trainable_params= 0
  all_param= 0
  for _, param in model.named_parameters():
    all_param+= param.numel()
    if param.requires_grad:
      trainable_params+= param.numel()

  print("trainable params:", trainable_params)
  print("all params:", all_param)
  print("trainable%:", (100 * trainable_params / all_param))
```

```python
config= peft.LoraConfig(
    r= 8,
    lora_alpha= 32,
    target_modules= ['query_key_value'],
    lora_dropout= 0.05,
    bias= "none",
    task_type= "CAUSAL_LM"
)

model= peft.get_peft_model(model, config)
print_trainable_parameters(model)
```

```python
data= load_dataset("Abirate/english_quotes")
```

```python
data= data.map(lambda x: tokenizer(x["quote"]), batched= True)
```

```python
tokenizer.pad_token= tokenizer.eos_token
```

```python
arguments= transformers.TrainingArguments(
    per_device_train_batch_size= 1,
    gradient_accumulation_steps= 4,
    warmup_steps= 2,
    max_steps= 5,
    learning_rate= 1e-5,
    fp16= True,
    logging_steps= 1,
    output_dir= 'outputs',
    optim="paged_adamw_8bit"
)
```

```python
data_collator= transformers.DataCollatorForLanguageModeling(tokenizer, mlm= False)
```

```python
data
```

```python
trainer= transformers.Trainer(
    model,
    arguments,
    train_dataset= data['train'],
    data_collator= data_collator
)
```

```python
model.config.use_cache= False
trainer.train()
```

```python
model_to_save= trainer.model.module if hasattr(trainer.model, 'module') else trainer.model
model_to_save.save_pretrained("outputs")
```

```python
lora_config= peft.LoraConfig.from_pretrained('outputs')
model= peft.get_peft_model(model, lora_config)
```

```python
text= "albert einstien"
device= "cuda:0"

inputs= tokenizer(text, return_tensors= 'pt').to(device)
outputs= model.generate(**inputs)
print(tokenizer.decode(outputs[0], skip_special_tokens= True))
```
